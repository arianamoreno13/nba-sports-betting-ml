### Concatinate the data into one large dataframe with 10 seasons of data and clean the data

```python
# create a combined dataframe with the 2011-2021 seasons
df_combined_2011_2021 = pd.concat([
    df_2011_2012,
    df_2012_2013,
    df_2013_2014,
    df_2014_2015,
    df_2015_2016,
    df_2016_2017,
    df_2017_2018,
    df_2018_2019,
    df_2019_2020,
    df_2020_2021
])
df_combined_2011_2021
```

```python
# Reset the index
df_combined_2011_2021 = df_combined_2011_2021.reset_index()
```

```python
# drop the "Unnamed: 0" column
df_combined_2011_2021 = df_combined_2011_2021.drop(
    columns=[
        'Unnamed: 0'
    ]
)
df_combined_2011_2021.head(1)
```

```python
# rename the "Home", "Away", "OU", "ML_Home", "ML_Away", and "Points" columns
df_combined_2011_2021 = df_combined_2011_2021.rename(
    columns={
        'Home':'Home_Team',
        'Away':'Away_Team',
        'OU':'Over_Under',
        'ML_Home':'Moneyline_Home_Team',
        'ML_Away':'Moneyline_Away_Team',
        'Points':'Game_Total_Points'
    }
)
df_combined_2011_2021.head(1)
```

```python
# check the data types of all the columns
df_combined_2011_2021.dtypes
```

```python
# change the 'Date' column to datetime data type
df_combined_2011_2021['Date'] = pd.to_datetime(df_combined_2011_2021['Date'])
df_combined_2011_2021
```

```python
# change the "Spread" column to float data type
df_combined_2011_2021['Spread'] = df_combined_2011_2021['Spread'].astype(float)
df_combined_2011_2021['Moneyline_Home_Team'] = df_combined_2011_2021['Moneyline_Home_Team'].astype(float)
df_combined_2011_2021['Moneyline_Away_Team'] = df_combined_2011_2021['Moneyline_Away_Team'].astype(float)
df_combined_2011_2021['Game_Total_Points'] = df_combined_2011_2021['Game_Total_Points'].astype(float)
df_combined_2011_2021['Win_Margin'] = df_combined_2011_2021['Win_Margin'].astype(float)
```

```python
# check data types are changed correctly
df_combined_2011_2021.dtypes
```

```python
# check for any null values
df_combined_2011_2021.isnull().sum()
```

```python
# check for any duplicated values
df_combined_2011_2021.duplicated().sum()
```
